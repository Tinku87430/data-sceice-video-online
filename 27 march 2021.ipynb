# Matplotlib Area and Scatter plot

```python
iris=sns.load_dataset('iris')
iris
```

```python
df=iris.drop(['species'],axis=1)
```

```python
df
```

```python
df.plot(kind='area')
```

```python
df.plot.scatter(x='sepal_length',y='petal_length')
```

```python
df.plot.scatter(x='sepal_length',y='petal_length',c='sepal_width')
```

```python
ax=df.plot.scatter(x='sepal_length',y='petal_length',label='length')
df.plot.scatter(x='sepal_width',y="petal_width",label="width",ax=ax,color='r')
```

# Seaborn Numerical data plot

1.Numerical Data Ploting
.relplot()
.Scatterplot()
.lineplot()

```python
tips=sns.load_dataset('tips')
```

```python
tips.tail()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,hue='smoker',style='time')
```

```python
sns.relplot(x='total_bill',y='tip',style='size',data=tips)
```

```python
sns.relplot(x='total_bill',y='tip',hue='size',data=tips,palette='ch:r=-0.5,l=0.95')
```

```python
sns.relplot(x='total_bill',y='tip',data=tips,size='size')
```

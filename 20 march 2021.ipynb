# Nan and None in pandas

```python
pd.Series([1,np.nan,2,None])
```

Create DataFrame with missing value

```python
raw_data={'first_name':['vikas',np.nan,'tina','hena','avi'],
         'last_name':['rajput',np.nan,'pal','raj','nash'],
         'age':['22',np.nan,23,24,25],
         'sex':['M',np.nan,'F','F','M'],
         'pretestscore':[4,np.nan,np.nan,2,3],
         'posttestscore':[25,np.nan,np.nan,62,70]}
df=pd.DataFrame(raw_data,columns=['first_name','last_name','age','sex','pretestscore','posttestscore'])
df
```

# Drop missing observations

```python
df_no_missing=df.dropna()
df_no_missing
```

# Drop rows where all cells in that raw in NA

```python
df_cleaned=df.dropna(how='all')
```

```python
df_cleaned
```

# Create a new column full of missing values

```python
df['location']=np.nan
```

```python
df
```

# Drop column if they only contain missing values

```python
df.dropna(axis=1,how='all')
```

```python
df.fillna(0)
```

# File in missing in pretestscore with the mean value of pretestscore

```python
df['pretestscore'].fillna(df['pretestscore'].mean(),inplace=True)
```

```python
df
```

# Fill in missing in posttestscore with each sex's mean value of posttestscore

```python
df["posttestscore"].fillna(df.groupby("sex")["posttestscore"].transform("mean"),inplace=True)
```

```python
df
```

# Select some raws but ignore the missing data points

```python
df[df['age'].notnull() & df["sex"].notnull()]
```

# Back-fill

```python
df.fillna(method='bfill')
```

# Forward fill

```python
df.fillna(method='ffill',axis=1)
```

# Hierarchical Indexing

# Create DataFrame

```python
row_data={'regiment':['nighthawks','nighthawks','nighthawks','nighthawks','dragoon','dragoon','dragoon','dragoon','scouts','scouts','scouts','scouts'],
         'company':['1st','1st','2nd','2nd','1st','1st','2nd','2nd','1st','1st','2nd','2nd'],
         'name':['miller','jackson','ali','milner','cooze','jackson','ryaner','soni','slaon','piger','riani','ali'],
         'preTestscore':[4,24,31,2,3,4,24,31,2,3,2,3],
         'postTestscore':[25,94,57,62,70,25,94,57,62,70,62,70]}

df=pd.DataFrame(row_data,columns=['regiment','company','name','preTestscore','postTestscore'])
df
```

# Set the hierarchical index but leave the columns inplace

```python
df.set_index(['regiment','company'],drop=False)
```

# Set the hierrarchical index to be by regiment and then by company

```python
df.set_index(['regiment','company'],inplace=True)
```

```python
df
```

# Swap the level in the index

```python
df.swaplevel('regiment','company')
```

# Summarize the result by regiment

```python
df.sum(level="regiment")
```
